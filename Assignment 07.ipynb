```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Salaries.csv')
df
```

#### Question 01 : Which employee has the highest TotalPayBenefits?

```python
max_index = np.argmax(df['TotalPayBenefits'])
```

```python
highest_paid_employee = df.iloc[max_index]
```

```python
print("Employee with highest TotalPayBenefits:")
print(highest_paid_employee)
```

#### Question 02: Drop ID, Notes, Agency and Status Columns

```python
df.drop(['Id', 'Notes', 'Agency', 'Status'], axis=1, inplace= True)
```

```python
df
```

#### Question 03: Total Number of Job Titles Contain POLICE

```python
df[df.JobTitle.str.contains('POLICE')]
```

#### Question 04: Top 10 most common job titles?

```python
top_10_jobs = df['JobTitle'].value_counts().head(10)
```

```python
print("Top 10 Most Common Job Titles:")
print(top_10_jobs.to_frame().reset_index().rename(columns={'index': 'JobTitle', 'JobTitle': 'Count'}))
```

#### Questio 05: What are the top 10 average TotalPay by JobTitle?

```python
avg_totalpay_by_job = df.groupby('JobTitle')['TotalPay'].mean()

top_10_avg_totalpay = avg_totalpay_by_job.sort_values(ascending=False).head(10)

print("Top 10 Job Titles by Average TotalPay:")
print(top_10_avg_totalpay.reset_index().rename(columns={'TotalPay': 'AverageTotalPay'}))
```

#### Question 06: How many unique job titles are there each year?

```python
unique_job_titles_per_year = df.groupby('Year')['JobTitle'].nunique()

unique_job_titles_df = unique_job_titles_per_year.reset_index()
unique_job_titles_df.columns = ['Year', 'UniqueJobTitleCount']

print("Unique Job Titles per Year:")
print(unique_job_titles_df.to_string(index=False))
```

#### Question 07: Find Average TotalPay Of All Employee Per Year

```python
avg_totalpay_per_year = df.groupby('Year')['TotalPay'].mean()

avg_totalpay_df = avg_totalpay_per_year.reset_index()
avg_totalpay_df.columns = ['Year', 'AverageTotalPay']

print("Average TotalPay per Year:")
print(avg_totalpay_df.to_string(index=False))
```

#### Question 08: How many unique employees worked each year?

```python
unique_employees_per_year = df.groupby('Year')['EmployeeName'].nunique()

unique_employees_df = unique_employees_per_year.reset_index()
unique_employees_df.columns = ['Year', 'UniqueEmployeeCount']

print("Unique Employees per Year:")
print(unique_employees_df.to_string(index=False))
```

#### Question 09: Which year had the most employees earning over 100,000 in TotalPay?

```python
high_earners = df[df['TotalPay'] > 100000]

high_earners_per_year = high_earners.groupby('Year')['EmployeeName'].nunique()

high_earners_df = high_earners_per_year.reset_index()
high_earners_df.columns = ['Year', 'EmployeesOver100K']

max_year = high_earners_df.loc[high_earners_df['EmployeesOver100K'].idxmax()]

print("Employees Earning Over 100,000 per Year:")
print(high_earners_df.to_string(index=False))

print(f"\n✅ Year with most employees earning over 100K: {int(max_year['Year'])} ({int(max_year['EmployeesOver100K'])} employees)")
```

#### Question 10: Which columns have missing values?

```python
missing_values = df.isnull().sum()

missing_columns = missing_values[missing_values > 0]

missing_columns_df = missing_columns.reset_index()
missing_columns_df.columns = ['Column', 'MissingValueCount']

print("Columns with Missing Values:")
print(missing_columns_df.to_string(index=False))
```
